### Step 4. Discover what is the mean age per occupation

```python
users[['age','occupation']].groupby('occupation').describe()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
users[['gender','occupation']].groupby('occupation').count()
```

```python
users[users['gender']=='M'][['gender','occupation']].groupby('occupation').count()
```

```python
round(100*(users[users['gender']=='M'][['gender','occupation']].groupby('occupation').count())/(users[['gender','occupation']].groupby('occupation').count()),2)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
que=users[['occupation','age']].groupby('occupation').describe()
que
```

```python
users.groupby('occupation').agg(['min', 'max','sum'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users[['occupation','gender','age']].groupby(['occupation','gender']).mean()
```
